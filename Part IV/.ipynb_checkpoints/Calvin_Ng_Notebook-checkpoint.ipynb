```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.formula.api as smf
import statistics as stat
```

```python
def cumulate_data(df, start_index, end_index):
    '''
    Given a start and end column, this function calculates
    the difference between each column and the one that
    proceeded it. It then stores these values in a list and
    the days since the start column in another list and returns
    both lists.
    :param start_index: The starting columnn.
    :param end_index: The ending column.
    :return: Two lists; one containing weeks and the other
    containing the summation of the columns of that week.
    '''
    x_vals = []
    y_vals = []
    
    # cumulate weekly data
    prev_col = df.iloc[:, start_index]
    prev_day = prev_col.sum()
    for i in range(start_index, end_index + 1):
        col = df.iloc[:, i]
        curr_day = col.sum()
        x_vals.append(i - start_index)
        y_vals.append(curr_day - prev_day)
        prev_day = curr_day
    
    return x_vals, y_vals
```

```python
nf = 100000

def normalize(dataset, population, normalization_factor):
    '''
    This function normalizes a set of values given a
    populationa nd normalization factor.
    
    :param dataset: An iterable structure containing the
    data we want to normalize.
    :param population: The population the data is describing.
    It is an integer.
    :param normalization_factor: An integer defining the
    noramlization factor; per 10,000, per 100,000, etc.
    '''
    normalized_data = []
    
    # avoid division by zero
    if population == 0:
        population = 1

    for data in dataset:
        normalized_data.append(data / population * normalization_factor)
        
    return normalized_data
```

```python
def label_graph(x_axis_label, y_axis_label, title, annotation=None, 
                xycoords='subfigure fraction', xy=(0.7, 0.8),
                ymax=None):
    '''
    This function labels the x-axis, y-axis, and title of a graph.
    If given, it will also place an annotation on that graph, by
    default in the upper-left corner.
    :param x_axis_label: a string containing the x-axis label.
    :param y_axis_label: a string containing the y-axis label.
    :param title: a string containing the title.
    :param annotation: a string containing an annotation.
    :param xycoords: a string containing what coordinate system should
    be used. See matplotlib annotations for documentation.
    :xparam xy: A tuple containing the x-y coordinates of the annotation,
    in the format specified by xycoords. See matplotlib annotations for
    documentation.
    '''
    plt.xlabel(x_axis_label)
    plt.ylabel(y_axis_label)
    plt.title(title)
    plt.legend()
    if annotation is not None:
        plt.annotate(annotation, xycoords = xycoords, xy = xy,
                    bbox=dict(boxstyle="round", fc="white", ec="grey", lw=0.5))
    if ymax is not None:
        plt.ylim(0, ymax * 1.5)
```

```python
def plot_regression(model, dataframe, x, label='past predictions'):
    '''
    Given a linear regression model, the x column inputs, and the
    dataframe these x column inputs are stored in, this function
    makes predictions, then plots those predictions as a line in
    the color red.
    :param linear_model: a statsmodel linear regression model.
    :param dataframe: a dataframe containing the x values.
    :param x: a string containing the name of the column in the
    dataframe containing the x values.
    :return: the predicted y values.
    '''
    predictions = model.predict(dataframe[x])
    plt.plot(dataframe[x], predictions, label=label)
    
    return predictions
```

```python
def calculate_rmse(y_vals, predicted_y_vals):
    '''
    This function calculates the rmse a linear model given its
    actual y values and its predicted y_values.
    :param y_vals: the actual y values of the data.
    :param predicted_y_vals: The predicted y values from the
    model.
    :return: the rmse.
    '''
    return np.sqrt(mean_squared_error(y_vals, predicted_y_vals))
```

```python
def predict_forecast(model, x, next_x_days):
    '''
    This function predicts y values over the next x days
    given a prediction model. It then graphs the forecast.
    :param model: the model used to predict y values.
    :param x: a series containing the past x values.
    :param next_x_days: an integer defining how many days
    to predict into the future.
    '''
    
    # make predictions
    x_pred = pd.Series([i for i in range(x.iloc[-1] + 1, next_x_days + x.iloc[-1] + 1)], dtype=int)
    x_pred = pd.DataFrame({'days': x_pred})
    predictions = model.predict(x_pred['days'])
    
    # plot predictions
    plt.plot(x_pred['days'], predictions, label='forecast')
    
    return predictions
```

```python
class CountyInfo:
    nf = 1000000
    
    def __init__(self, name, population, daily_cases, daily_deaths):
        self.name = name.replace(' ', '_')
        self.population = population
        self.daily_cases_norm = normalize(daily_cases, self.population, nf)
        self.daily_deaths_norm = normalize(daily_deaths, self.population, nf)
        self.mean_cases = stat.mean(self.daily_cases_norm)
        self.mean_deaths = stat.mean(self.daily_deaths_norm)
```

Similar to the group notebook for this stage, we will create linear and polynomial models for North Carolina. The process will follow very closely to the group notebook, and we will reuse much of that code.

We'll first create a pandas dataframe from the notebook and pull new daily cases and deaths into Pandas Series along with days, then we'll create models, graph them, create a forecast, calculate their RMSE, and calculate their confidence intervals.

```python
# read in dataset into a dataframe and drop non-NC rows
base = pd.read_csv('../Data/base_set_joined.csv')
nc_df = base.drop(base[base['State'] != 'NC'].index)
nc_population = nc_df['population'].sum()
nf = 1000000
```

```python
# get start and end indices
start_index = base.columns.get_loc("cases 2020-01-22")
end_index = base.columns.get_loc("cases 2023-01-16")

# get cases data
days, nc_cases = cumulate_data(nc_df, start_index, end_index)

# get start and end indices
start_index = base.columns.get_loc("deaths 2020-01-22")
end_index = base.columns.get_loc("deaths 2023-01-16")

# get deaths data
_, nc_deaths = cumulate_data(nc_df, start_index, end_index)

# convert to Series
days = pd.Series(days)
nc_cases_normalized = pd.Series(normalize(nc_cases, nc_population, nf))
nc_deaths_normalized = pd.Series(normalize(nc_deaths, nc_population, nf))
nc_cases = pd.Series(nc_cases)
nc_deaths = pd.Series(nc_deaths)

# construct dataframe
df = pd.DataFrame({'days': days, 'nc_cases': nc_cases, 'nc_deaths': nc_deaths})
```

```python
# plot scatterplot
df.plot(kind = 'scatter', x = 'days', y = 'nc_cases', label='observed data', c='darkseagreen', s=5)

# create and plot linear regression model
nc_cases_lm = smf.ols(formula = 'nc_cases ~ 1 + days', data = df).fit()
nc_cases_lm_predictions = plot_regression(nc_cases_lm, df, 'days')

# plot forecast
ymax = predict_forecast(nc_cases_lm, df['days'], 365).max()

# calculate the rmse and confidence interval
nc_cases_lm_rmse = calculate_rmse(df['nc_cases'], nc_cases_lm_predictions)
nc_cases_lm_ci = nc_cases_lm.conf_int()
annotation = 'rmse: %.2f\nslope cf: %.2f, %.2f' % (nc_cases_lm_rmse, nc_cases_lm_ci.iloc[1, 0], nc_cases_lm_ci.iloc[1, 1])

# label graph
label_graph('Days Since the First Case', 'New Cases', 'Linear Regression Model of NC Cases', 
            annotation, ymax=ymax)
```

### North Carolina Models Analysis

In general, similar to the models for the United States as a whole, none of these models particularly fit our data well. The issue is that the data is so eclectic that neither cases or deaths follow a general upward or downward trend. As a result, it's difficult to fit a model to the data.

The polynomial models capture the peaks of cases and deaths over the duration of the pandemic. Like with the country analysis, the deaths seem to peak earlier, around day 500, than the cases, which peaks around day 600.

However, the polynomial models are not ideal for future forecasts due to the fact that they begin to predict negative cases and deaths. As a result, the linear models are likely better for this task. While they do not capture the true nature of the data, they at least predict positive values.

In order to attain a better model, we would likely have to reduce the time frame we are looking at. For example, a timeframe of a few months would likely give us a much better model, however, that model would likely only be able to make accurate predictions up to a certain point, no more than a month in the future.

## Section III: County Models

The next section covers the models for counties. 

First, we'll start by actually identifying the top five counties with the highest cases and highest deaths. To do this, we'll normalize the data and calculate the average number of cases and deaths for each county, then choose the counties with the highest averages.

After we have normalized the data, we can then add each county's cases and deaths as columns in our dataframe. Once our dataframe has been populated, we can proceed to train the models and plot the graphs as we did for North Carolina as a whole.

```python
counties_info = []

# get start and end indices
start_index_cases = nc_df.columns.get_loc("cases 2020-01-22")
end_index_cases = nc_df.columns.get_loc("cases 2023-01-16")
start_index_deaths = nc_df.columns.get_loc("deaths 2020-01-22")
end_index_deaths = nc_df.columns.get_loc("deaths 2023-01-16")

first_index = nc_df.index[0]

# populate counties info list
for i in range(first_index, first_index + nc_df.shape[0]):
    curr_row = nc_df.loc[[i]]
    county_cases = cumulate_data(curr_row, start_index_cases, end_index_cases)[1]
    county_deaths = cumulate_data(curr_row, start_index_deaths, end_index_deaths)[1]
    county_name = curr_row.iloc[0]['County Name']
    county_population = curr_row.iloc[0]['population']
    county = CountyInfo(county_name, county_population, county_cases, county_deaths)
    counties_info.append(county)
```

```python
# calculate counties with the highest cases and highest deaths
# for cases, we pull top 6 and drop the first element because the top county is 'statewide unallocated', which isn't a county
highest_cases = [county for county in sorted(counties_info, key=lambda x: x.mean_cases, reverse=True)[:6]] 
del highest_cases[0]
highest_deaths = [county for county in sorted(counties_info, key=lambda x: x.mean_deaths, reverse=True)[:5]]

# print results
print('Counties With the Highest Cases:')
for county in highest_cases:
    print('%-20s%10.2f' % (county.name, county.mean_cases))
    df[county.name + '_cases'] = county.daily_cases_norm
print('\nCounties With the Highest Deaths:')
for county in highest_deaths:
    print('%-20s%10.2f' % (county.name, county.mean_deaths))
    df[county.name + '_deaths'] = county.daily_deaths_norm
```

```python
# plot scatterplot
df.plot(kind = 'scatter', x = 'days', y = 'Robeson_County_cases', label='observed data', c='darkseagreen', s=5)

# create and plot linear regression model
Robeson_County_cases_lm = smf.ols(formula = 'Robeson_County_cases ~ 1 + days', data = df).fit()
Robeson_County_cases_lm_predictions = plot_regression(Robeson_County_cases_lm, df, 'days')

# plot forecast
ymax = max(predict_forecast(Robeson_County_cases_lm, df['days'], 365).max(), Robeson_County_cases_lm_predictions.max())

# calculate the rmse and confidence interval
Robeson_County_cases_lm_rmse = calculate_rmse(df['Robeson_County_cases'], Robeson_County_cases_lm_predictions)
Robeson_County_cases_lm_ci = Robeson_County_cases_lm.conf_int()
annotation = 'rmse: %.2f\n' % Robeson_County_cases_lm_rmse
annotation += 'slope cf: %.2f, %.2f' % (Robeson_County_cases_lm_ci.iloc[1, 0], Robeson_County_cases_lm_ci.iloc[1, 1])

# label graph
label_graph('Days Since the First Case', 'New Cases per Million People', 'Linear Regression Model of Robeson County Cases', 
            annotation, xy=(0.15, 0.8), ymax=ymax)
```

### Counties Analysis

Similar to state and nation models, these models do not appear to be very good at capturing the nature of the data. However, we can still learn some things from the regardless.

Let's begin with the linear models. I anticipated that counties with a higher average of cases would also have steeper slopes, but this was not always the case. As an example, despite Robeson County having a higher average of cases than Yadkin County, Yadkin County has a steeper slope as determined by the 95% confidence interval. However, Robeson County's intercept is greater than Yadkin County's intercept. 

The point is that despite all of these counties having high numbers of cases, the nature of the pandemic differs from county to county. In Robeson County, we can infer from a high intercept and lower slope that the county was hit hard in the beginning of the pandemic, but the increase in number of cases has been slow. In Yadkin County, the county began with fewer cases, but the spread was more prolific and has resulted in a more rapid increase in the rise of cases.

For the most part, the polynomial models look very similar; for starters, they all appear to peak around 500 cases per million people. What day they peak slightly differs, although most are in the range of day 600 to 700.

## Section IV: Hypothesis Testing

In this section, we will test the following hypothesis: there a difference in the correlations of cases to employment between North Carolina and the nation as a whole in the year 2021. 

In the previous section, we calculated correlations between covid cases and some of the variables from the employment enrichment dataset. We did this for the whole nation and over quarter three of 2022. Of the variables we tested, the correlation between cases and employment as a percentage of the population working was the strongest.

So, I want to analyze the correlation in North Carolina and determine whether it differs from that of the United States as a whole. We will be using data from 2021 since it will give us greater consistency by pulling data from an entire year instead of a single quarter. Furthermore, I have already taken care of the merging required all the way back in part one, so it makes sense to actually take advantage of it.

The type of hypothesis test we will conduct is a Fisher's z-test. While it is not included in our lecture notes, I did some research and determined that this hypothesis test is ideal and matches the description of what I am trying to do. 

To begin with, we will first calculate the correlations between cases and employment for North Carolina and the United States as a whole. The process will be similar to that in stage three, where I did this computation for the nation. Once the correlations have been calculated, we will then move on to the hypothesis test.

```python
# read joined dataset
employment_df = pd.read_csv('../Data/employment_base_merged.csv', low_memory=False)
employment_df.head()
```

```python
states = employment_df['State'].unique()
nation_df = pd.DataFrame(columns=['state', 'employment'])

# populate states data
for state in states:
    # select current state
    state_df = employment_df.drop(employment_df[employment_df['State'] != state].index)
    
    # store state population
    population = state_df['population'].sum()

    # get start and end indices
    start_index = state_df.columns.get_loc("cases 2021-01-01")
    end_index = state_df.columns.get_loc("cases 2021-12-31")

    # calculate new cases
    _, new_cases = cumulate_data(state_df, start_index, end_index)   
    
    # store mean of new cases
    cases_normalized = normalize(new_cases, population, nf)
    cases_mean = np.mean(cases_normalized)
    
    # get and store employment as a percentage
    employment = state_df['Annual Average Employment'].sum()
    employment_percentage = employment / population

    # add a row for the current state
    new_row = pd.DataFrame({'state': [state], 'employment': [employment_percentage], 'cases': cases_mean})
    nation_df = pd.concat([nation_df, new_row], ignore_index=True)

nation_df
```
